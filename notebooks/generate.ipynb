# Generate Data

combined data from original raw:

https://www.kaggle.com/datasets/psparks/instacart-market-basket-analysis/data


```python
import pandas as pd
```

```python
PATH_DATA = "../../../dt/instacart-data/"
```

## Get Data

```python
df_cart = pd.read_csv(PATH_DATA+'order_products__prior.csv')
df_products = pd.read_csv(PATH_DATA+'products.csv',usecols=["product_id","department_id","product_name"])
df_orders = pd.read_csv(PATH_DATA+'orders.csv',usecols=["order_id","user_id","order_number"])
df_department = pd.read_csv(PATH_DATA+'departments.csv')
```

```python
df_orders
```

```python
df_m1 = pd.merge(df_cart, df_products,on="product_id",how="left")
df_m1 = pd.merge(df_m1, df_department, on="department_id", how="left")
df_m1 = pd.merge(df_m1, df_orders, on="order_id", how="left")

df_m1.sample(10)
```

## Summarize it

```python
# df_m1.groupby(["order_id","user_id"]).product_id.count()
df_m2 = df_m1.groupby(["order_id","order_number","user_id","department"]).product_id.count()
df_m2 = df_m2.unstack().fillna(0).reset_index()
```

```python
df_m2.to_csv(PATH_DATA+"basket_data.csv",index=False)
```
